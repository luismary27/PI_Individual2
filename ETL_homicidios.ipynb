```python
#Modificamos los nombres de las columnas pos x/pos y que segun el diccionario es Longuitud y latitud
homi_hechos = homi_hechos.rename(columns={'pos x': 'LONGITUD', 'pos y': 'LATITUD'})
#Modificar los nombres de la fecha
homi_hechos = homi_hechos.rename(columns={'AAAA': 'AÑO', 'MM': 'MES', 'DD': 'DIA'})
homi_hechos
```

```python
#Revisamos el df homicidios victimas
homi_victimas
```

```python
#Eliminamos la rebundacia de datos 
homi_victimas = homi_victimas.drop(['FECHA', 'AAAA', 'MM', 'DD','VICTIMA'], axis=1)
```

```python
#Reemplazamos los valores nulos
homi_victimas.replace(['SD','sd','Sd'], np.nan, inplace=True)
```

```python
null = homi_victimas[homi_victimas.isnull().any(axis=1)]
null
```

```python
#En las columna de ROL se deja en blanco los valores NaN 
homi_victimas['ROL'] = homi_victimas['ROL'].fillna('')
#En la columna SEXO Colocamos NO DEFINIDO
homi_victimas['SEXO'] = homi_victimas['SEXO'].fillna('NO DEFINIDO')
#En la columna EDAD los remplazamos por 0
homi_victimas['EDAD'] = homi_victimas['EDAD'].fillna(0)
#En la columna FECHA FALLECIMIENTO reemplazamos por NO FALLECIDO
homi_victimas['FECHA_FALLECIMIENTO'] = homi_victimas['FECHA_FALLECIMIENTO'].fillna('NO FALLECIDO')

homi_victimas
```

```python
#Realizamos este cambio de nombre para poder hacer el merge
homi_victimas= homi_victimas.rename(columns={'ID_hecho': 'ID'})
homi_victimas
```

```python
#Realizamos el merge 
Homicidos_df = homi_hechos.merge(homi_victimas, on='ID', how='left')
Homicidos_df
```

```python
#Cambiamos el nombre de la columa rol ya que no queda claro
Homicidos_df = Homicidos_df.rename(columns={'ROL': 'ROL_VICTIMA'})
```

```python
#Cambiamos el tipo de dato de la columna "FECHA FALLECIMIENTO"
Homicidos_df['FECHA_FALLECIMIENTO'] = pd.to_datetime(Homicidos_df['FECHA_FALLECIMIENTO'], errors='coerce')
Homicidos_df['FECHA_FALLECIMIENTO'] = Homicidos_df['FECHA_FALLECIMIENTO'].dt.strftime('%Y-%m-%d')
Homicidos_df['FECHA_FALLECIMIENTO'] = Homicidos_df['FECHA_FALLECIMIENTO'].astype('datetime64[ns]')
Homicidos_df
```

```python
#Lo guardamos en un archivo csv para el EDA
Homicidos_df.to_csv('Datasets/Homicidios_clean.csv', index=False)
```
