```python
#Graficamos outliers
columnas= ['N_VICTIMAS','EDAD']
colores= ['red', 'blue']

plt.figure(figsize=(15, 5))

for i, column in enumerate(columnas):
    plt.subplot(2, 5, i+1)  
    sns.boxplot(y=homicidios[column], color=colores[i])
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
#AÑO
siniestros_anio = homicidios.groupby(homicidios['AÑO']).size()

plt.plot(siniestros_anio.index, siniestros_anio, label='Año', marker='o', alpha=0.5, color='Red')
plt.title('Siniestros Fatales por Año')
plt.xlabel('Año')
plt.ylabel('Número de Siniestros Fatales')
plt.grid(True)
```

```python
print("\n\033[1mCantidad de fallecidos por año:\033[0m")
siniestros_anio= homicidios['AÑO'].value_counts().sort_index(ascending=True)
siniestros_anio
```

El año 2018 se destaca como el de mayor cantidad de siniestros dentro del periodo analizado.  
2018-2020: Se observa una disminución significativa en el número de siniestros. La pandemia podría estar relacionada con el aumento de 2021, pero no es la única causa posible.  
El año 2021 muestra un repunte, sin superar el pico del 2018. 

```python
#MES
siniestros_mes=homicidios.groupby(homicidios['MES']).size()
plt.figure(figsize=(12, 6))
plt.fill_between(siniestros_mes.index, siniestros_mes.values, color='skyblue')
plt.xticks(rotation=0)
plt.xlabel('Mes')
plt.ylabel('Total siniestros')
plt.title('Total siniestros por mes')
plt.show()
```

```python
print("\n\033[1mCantidad de fallecidos por mes:\033[0m")
siniestros_mes= homicidios['MES'].value_counts().sort_index(ascending=True)
siniestros_mes
```

Si bien Agosto, Noviembre y Diciembre presentan la mayor cantidad de siniestros (67, 67 y 81 respectivamente), la diferencia con respecto al resto del año no es significativa.

```python
#SEXO
siniestros_genero = homicidios['SEXO'].value_counts()
porcen_genero = (siniestros_genero / siniestros_genero.sum()) * 100

plt.figure(figsize=(5, 5))
plt.pie(porcen_genero, labels=porcen_genero.index, autopct='%1.1f%%', startangle=90, colors=['red', 'blue','green' ])
plt.title('Distribución de Género en Siniestros Fatales')
plt.show()
```

```python
print("\n\033[1mCantidad de sinietro por generos:\033[0m")
siniestros_genero= homicidios['SEXO'].value_counts().sort_index(ascending=True)
siniestros_genero
```

Se puede observar que los hombres (MASCULINOS) son los mas propensos a sufrir siniestros fatales

```python
#COMUNA
cant_por_comuna = homicidios['COMUNA'].value_counts()
cant_por_comuna_porcentaje = cant_por_comuna.values / cant_por_comuna.sum() * 100

plt.figure(figsize=(10, 5))
plt.barh(cant_por_comuna.index, cant_por_comuna.values)
plt.yticks(rotation=0)
plt.xlabel('Cantidad')
plt.ylabel('Comuna')
plt.title('Cantidad de siniestros por Comuna')
plt.show()
```
